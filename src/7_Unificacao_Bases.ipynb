### 3. Realizar join/merge + validar junção

```python
# Realizar o merge com indicador de origem
df_joined = pd.merge(
    df_enem,
    df_censo,
    left_on=['CAT_DEPENDENCIA_ADM_ESC', 'CAT_CO_MUNICIPIO_ESC', 'CAT_LOCALIZACAO_ESC'],
    right_on=['CHA_DEPENDENCIA', 'CHA_CO_MUNICIPIO', 'CHA_LOCALIZACAO'],
    how='outer',
    indicator=True
)

# Linhas só no ENEM
df_somente_enem = df_joined[df_joined['_merge'] == 'left_only']

# Linhas só no Censo
df_somente_censo = df_joined[df_joined['_merge'] == 'right_only']

# Linhas que estão em ambos
df_ambos = df_joined[df_joined['_merge'] == 'both']

# Visualizar quantidades
print('Somente ENEM:', len(df_somente_enem))
print('Somente Censo:', len(df_somente_censo))
print('Ambos:', len(df_ambos))
```

```python
# Realizar um join entre as duas bases com as três colunas como chave
df_joined = pd.merge(
    df_enem,
    df_censo,
    left_on=['CAT_DEPENDENCIA_ADM_ESC', 'CAT_CO_MUNICIPIO_ESC', 'CAT_LOCALIZACAO_ESC'],
    right_on=['CHA_DEPENDENCIA', 'CHA_CO_MUNICIPIO', 'CHA_LOCALIZACAO'],
    how='inner'
)
```

```python
# Remover colunas utilizadas para o join apenas do Censo
df_joined = df_joined.drop(columns=['CHA_DEPENDENCIA', 'CHA_CO_MUNICIPIO', 'CHA_LOCALIZACAO'])
```
